```python
import pandas as pd
```

```python
concentration = pd.read_csv('../../data/2018/18_concentrations.csv')
labels = pd.read_csv('../../data/2018/18_yellow_labels.csv')
pit = pd.read_csv('../../data/2018/18_time_since_pit.csv')
weather = pd.read_csv('../../data/weather_data/2018_weather.csv', sep=';')
speed = pd.read_csv('../../data/2018/18_avg_speed.csv')
```

```python
max_concentration = pd.DataFrame(columns=['Time', 'Max Concentration'])
temp_max = 0
for index, row in concentration.iterrows():
    sector = row[1]
    if sector == 1:
        temp_max = 0
    if row[2] > temp_max:
        temp_max = row[2]
    if sector == 13:
        max_concentration.loc[index] = [row[0], temp_max]
```

```python
start_time = weather['TIME_UTC_SECONDS'].iloc[0]

with open('../../data/2018/18_weather_mod.csv', 'w') as f:
    f.write('Time,Air Temp,Track Temp,Wind Speed\n')
    for index, row in weather.iterrows():
            if row['TRACK_TEMP'] != 32:
                temp = row['TRACK_TEMP']
            else:
                row['TRACK_TEMP'] = temp
        
            f.write("%s,%s,%s,%s\n"%((row[0]-start_time)//20 * 20, row[2],row[3], row[6]))
            f.write("%s,%s,%s,%s\n"%((row[0]-start_time+20)//20 * 20,row[2],row[3], row[6]))
            f.write("%s,%s,%s,%s\n"%((row[0]-start_time+40)//20 * 20,row[2],row[3], row[6]))
weather_mod = pd.read_csv('../../data/2018/18_weather_mod.csv')
```

```python
agg = max_concentration.merge(pit, how='inner', on='Time')
agg = agg.merge(weather_mod, how='inner', on='Time')
# agg = agg.merge(speed, how='inner', on='Time')
agg = agg.merge(labels, how='inner', on='Time')
agg
```

```python
agg.to_csv('../../data/2018/18_agg_data.csv', index=False)
```

```python
times = [20580, 21760, 25560, 29380, 38320]
times = [t//60*60 for t in times]
MAXTIME = 43200
with open('../../data/2018/18_flag_times.csv', 'w') as f:
    f.write('Time,Flag\n')
    for i in range(0,MAXTIME,60):
        if i+120 in times:
            f.write('%s,1\n'%(i))
            f.write('%s,1\n'%(i+60))
            f.write('%s,1\n'%(i+120))
            i = i + 120
        else:
            f.write('%s,0\n'%(i))

    
```

```python
labels = pd.read_csv('../../data/2018/18_flag_times.csv')
risk = pd.read_csv('../../data/2018/18_risk_score.csv')
sus = pd.read_csv('../../data/2018/18_sus_deriv.csv')
agg = risk.merge(sus, how='inner', on='Time')
agg = agg.merge(labels, how='inner', on='Time')
agg.to_csv('../../data/2018/18_agg_score.csv', index=False)
```
